AIM-To apply principal componenet analysis for dimensionality reduction and visulaize the transformation feature space

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(df)
print(x_scaled)
```

```python
pca = PCA(n_components=2)
x_pca = pca.fit_transform(x_scaled)
print(x_pca)
```

```python
print("explain variance")
pca.explained_variance_ratio_
```

```python
plt.figure(figsize=(8, 6))
scatter = plt.scatter(x_pca[:, 0], x_pca[:, 1], c=iris.target, cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Iris Dataset')
legend = plt.legend(*scatter.legend_elements(), title='Species')
for i, species in enumerate(iris.target_names):
    legend.get_texts()[i].set_text(species)
plt.grid(True)
plt.show()
```

```python
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)

plt.figure(figsize=(8, 5))
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o', linestyle='--')
plt.title('Cumulative Explained Variance by Principal Component')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance Ratio')
plt.grid(True)
plt.axhline(y=0.95, color='r', linestyle=':', label='95% Cumulative Variance') # Example: show line for 95%
plt.legend()
plt.show()
```
